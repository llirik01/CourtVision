### Pearson correlation

```python
from scipy import stats
```

```python
def correlation(df):
    """
    Compute Pearson correlation coefficient between each numeric feature and the target variable.
    Remove features with a high p-value (insignificant correlation).

    :param df: pd.DataFrame - Input dataset containing numeric features and the target variable.
    :return: pd.DataFrame - DataFrame with correlation coefficients, p-values, and feature status.
    """
    # Lists to store correlation coefficient, p-value, column names, and feature status
    corr_coef, pear_value, indexx, status = [], [], [], []

    # Iterate through numeric columns
    for col in df.mode(numeric_only = True).columns.tolist():

        # Compute Pearson correlation coefficient and p-value
        p_coef, p_value = stats.pearsonr(df[col], df['result'])
        corr_coef.append(round(p_coef, 5))
        pear_value.append(round(p_value, 5))
        indexx.append(col)

        # Determine whether to keep or remove the feature based on p-value threshold
        if p_value > 0.1:
            df.drop(col, axis = 1, inplace = True)
            status.append('will be deleted')
        else:
            status.append('stays')
            
    # Create a DataFrame summarizing correlation results
    dictt = {'correlation coef': corr_coef, 'p-value': pear_value, 'status': status}
    frame = pd.DataFrame(dictt, index = indexx)
    
    return frame
```

# Models

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score, log_loss, precision_score, recall_score
```

```python
# Separate the target variable Y and the features X
Y = data_atp['result']
X = data_atp.drop(['result'], axis=1)
Y.value_counts()
```

```python
# Define the test sizes and minimization methods for the optimization process
size = [round(i, 2) for i in np.arange(0.1, 0.25, 0.05)]  # Different test set sizes
minimization_metod = ['lbfgs', 'liblinear', 'newton-cg', 'sag', 'saga']  # Different solvers for logistic regression

for k in size:
    # Split the data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = k, random_state=42)
    print(f'\ntest_size = {k}:\n')
    for i in minimization_metod:
        logreg = LogisticRegression(solver=i)  # Initialize the logistic regression model with the chosen solver
        logreg.fit(X_train, y_train)  # Fit the model with the training data
        y_pred = logreg.predict(X_test)  # Make predictions on the test set
        print(f'solver = {i}, accuracy = {round(accuracy_score(y_test, y_pred), 5)}')
```

Best params: `test_size = 0.1`; `solver = newton-cg`

```python
# # Performing GridSearch
# # Split the dataset into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.1, random_state=42)

# Initialize the logistic regression model
logreg = LogisticRegression(solver='newton-cg')

grid = GridSearchCV(logreg, param_grid = {'fit_intercept': [True, False], \
                                          'C': [i for i in range(1, 6)], \
                                          'penalty': ['l2', None]})  # Solver newton-cg supports only 'l2' or 'None' penalties.

grid.fit(X_train, y_train)
best = grid.best_params_
best   # Show the best parameters
```

Best params: {'C': 5, 'fit_intercept': True, 'penalty': None}


```python
# Refining logistic regression parameters found using GridSearchCV
logreg = LogisticRegression(solver='newton-cg', C=5, penalty=None)

logreg.fit(X_train, y_train)  # Train the model using the training data

y_pred = logreg.predict(X_test)  # Make predictions on the test set
round(accuracy_score(y_test, y_pred), 5)
```

```python
def overfitting(model, X_test, y_test, X_train, y_train):
    """
    Evaluate model performance on test and training sets to check for overfitting.

    :param model: sklearn model - Trained model to evaluate.
    :param X_test: pd.DataFrame - Test set features.
    :param y_test: pd.Series - Test set target variable.
    :param X_train: pd.DataFrame - Training set features.
    :param y_train: pd.Series - Training set target variable.
    :return: pd.DataFrame - DataFrame with evaluation metrics for test and train sets.
    """
    # Lists to store evaluation metrics for test and train sets
    best_inter, best_C, acc, r_auc, f1, prec, rec, logl = ([] for i in range(8))

    # Scoring metrics to evaluate the model
    scoring = ['accuracy', 'roc_auc', 'f1', 'precision', 'recall', 'neg_log_loss']
    metrics_l = [acc, r_auc, f1, prec, rec, logl]

    # Evaluate the model on the test set using cross-validation
    for i in zip(metrics_l, scoring):
        el = cross_val_score(model, X_test, y_test, cv = 5, scoring = i[1])

        # Convert negative log loss to positive for better readability
        if i[1] == 'neg_log_loss':
            el = -1 * el

        i[0].append(el.mean())

    # Evaluate the model on the training set using cross-validation
    for k in zip(metrics_l, scoring):
        el = cross_val_score(model, X_train, y_train, cv = 5, scoring = k[1])

        if k[1] == 'neg_log_loss':
            el = -1 * el

        k[0].append(el.mean())

    # A dictionary to store test and train evaluation results
    dictt = {
             'Accuracy': acc, 
             'Precision': prec,
             'Recall': rec,
             'F1 score': f1,
             'Roc auc': r_auc,
             'Log loss': logl
            }

    result = pd.DataFrame(dictt, index = ['test', 'train'])

    return result.round(5)
```

```python
overfitting(logreg, X_test, y_test, X_train, y_train)
```

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
# Define the test sizes and minimization methods for the optimization process
size = [round(i, 2) for i in np.arange(0.1, 0.25, 0.05)]  # Different test set sizes

for k in size:
    # Split the dataset into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = k, random_state=42)
    print(f'\ntest_size = {k}:\n')
    # Iterate through different numbers of trees in the Random Forest
    for i in range(100, 300, 50):
        rfc = RandomForestClassifier(n_estimators=i)  # Initialize the Random Forest model
        rfc.fit(X_train, y_train)  # Train the model on the training data
        y_pred = rfc.predict(X_test)  # Make predictions on the test set
        print(f'n_estimators = {i}, accuracy = {round(accuracy_score(y_test, y_pred), 5)}')
```

Best params: `test_size = 0.1`; `n_estimators = 150`

```python
# # Performing GridSearch
# # Split the dataset into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.1, random_state=42)

rfc = RandomForestClassifier(n_estimators=150)  # Initialize the Random Forest model

grid = GridSearchCV(rfc, param_grid = {'criterion': ['gini', 'entropy', 'log_loss'], \
                                       'min_samples_leaf': [i for i in range(1, 4)], \
                                       'max_features': ['sqrt', 'log2', None]})

grid.fit(X_train, y_train)
best = grid.best_params_
best  # Show the best parameters
```

Best params: {'criterion': 'log_loss', 'max_features': None, 'min_samples_leaf': 2}

```python
# Refining Random Forest parameters found using GridSearchCV
rfc = RandomForestClassifier(n_estimators=150, criterion='log_loss', max_features=None, min_samples_leaf=2)

rfc.fit(X_train, y_train)  # Train the model using the training data

y_pred = rfc.predict(X_test)  # Make predictions on the test set
round(accuracy_score(y_test, y_pred), 5)
```

```python
overfitting(rfc, X_test, y_test, X_train, y_train)
```

## XGBoost

```python
import xgboost as xgb
```

```python
# Define the test sizes and minimization methods for the optimization process
size = [round(i, 2) for i in np.arange(0.1, 0.25, 0.05)]  # Different test set sizes

for k in size:
    # Split the dataset into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = k, random_state=42)
    print(f'\ntest_size = {k}:\n')
    # Iterate through different numbers of estimators in the XGBoost model
    for i in range(100, 300, 50):
        xgbc = xgb.XGBClassifier(n_estimators=i, objective='binary:logistic', eval_metric='logloss', use_label_encoder=False)
        xgbc.fit(X_train, y_train)
        y_pred = xgbc.predict(X_test)
        print(f'n_estimators = {i}, accuracy = {round(accuracy_score(y_test, y_pred), 5)}')
```

Best params: `n_estimators = 100`; `test_size = 0.15`

```python
# # Performing GridSearch
# # Split the dataset into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.15, random_state=42)

# Initialize the XGBoost model
xgbc = xgb.XGBClassifier(n_estimators=100, objective='binary:logistic', eval_metric='logloss', use_label_encoder=False)

grid = GridSearchCV(xgbc, param_grid = {'learning_rate': [0.01, 0.05, 0.1],
                                        'max_depth': [i for i in range(3, 9)],
                                        'reg_lambda': [1, 1.5, 2],
                                        'reg_alpha': [0, 0.1, 0.5]})

grid.fit(X_train, y_train)
best = grid.best_params_
best
```

Best params: {'learning_rate': 0.05, 'max_depth': 6, 'reg_alpha': 0.5, 'reg_lambda': 1.5}

```python
# Refining XGBoost parameters found using GridSearchCV
xgbc = xgb.XGBClassifier(n_estimators=100, objective='binary:logistic', 
                         eval_metric='logloss', use_label_encoder=False, learning_rate=0.05, 
                         reg_alpha=0.5, reg_lambda=1.5)

xgbc.fit(X_train, y_train)  # Train the model using the training data

y_pred = xgbc.predict(X_test)  # Make predictions on the test set
round(accuracy_score(y_test, y_pred), 5)
```

```python
overfitting(xgbc, X_test, y_test, X_train, y_train)
```
