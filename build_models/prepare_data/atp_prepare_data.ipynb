```python
def transform_seed(value):
    """
    Transform values in the 'winner_seed' or 'loser_seed' columns.

    'Q', 'LL', and 'WC' remain unchanged.
    Numeric values are converted to 'S' (Seed).
    NaN and 'A' are converted to 'UN' (Unseed).

    :param value: str or NaN - Value from the 'winner_seed' or 'loser_seed' columns
    :return: str - Transformed value ('S', 'UN', 'WC', 'Q', 'LL', or original value)
    """
    if pd.isna(value) or value == 'A':
        return 'UN'
    elif value in ['WC', 'Q', 'LL']:
        return value
    else:
        try:
            float(value)
            return 'S'
        except ValueError:
            return value

data_atp['winner_seed'] = data_atp['winner_seed'].apply(transform_seed)
data_atp['winner_seed'].unique()
```

```python
# Use a regular expression to find letters (without digits and symbols)
# The pattern [a-zA-Z] matches any letter of the alphabet (both uppercase and lowercase)
mask = data_atp['score'].str.contains(r'[a-zA-Z]')

# Select all rows where the 'score' column contains letters
records_with_letters = data_atp[mask]

records_with_letters.tail(3)
```

```python
def extract_letters(value):
    """
    Extract words (letters) from a string.

    :param value: str or any - The value from the 'score' column in the dataframe.
    :return: list - A list of letter sequences found in the 'value'.
    """
    return re.findall(r'[a-zA-Z]+', str(value)) 

# Apply the function to the 'score' column and concatenate all results into one list
letters_in_scores = data_atp['score'].apply(extract_letters).sum()

# Get the unique letter marks from the concatenated list
unique_letters = pd.Series(letters_in_scores).unique()

unique_letters
```

```python
#      
def remove_bracketed_numbers(score):
    """
    Remove numbers inside parentheses.

    :param score: str - The match score as a string, potentially containing numbers inside parentheses.
    :return: str - The score string with any numbers inside parentheses removed.
    """
    return re.sub(r'\(\d+\)', '', score)

# Apply the function to the 'score' column to remove bracketed numbers from all scores
data_atp['score'] = data_atp['score'].apply(remove_bracketed_numbers)

data_atp['score'].tolist()[:10]
```

```python
#       
def calculate_games(score):
    """
    Calculate the number of games and tiebreaks from a score string

    :param score: str - The match score string, e.g., '6-3 7-5 6-7 6-4'.
    :return: pd.Series - A series with the following values:
             - w_games: Total games won by the winner
             - l_games: Total games won by the loser
             - total_games: Total number of games in the match
             - p1_tiebreaks_w: Number of tiebreaks won by the winner
             - p2_tiebreaks_w: Number of tiebreaks won by the loser
             - tiebreaks: Total number of tiebreaks in the match
    """
    # Split the score into sets and each set into won/lost games
    sets = re.findall(r'\d+-\d+', score)
    
    w_games = 0
    l_games = 0
    w_tiebreaks_w = 0
    l_tiebreaks_w = 0

    for set_score in sets:
        # Split the score for each set into winner's and loser's games
        w_set, l_set = map(int, set_score.split('-'))

        # Add the number of games won by each player
        w_games += w_set
        l_games += l_set

        # Check for tiebreaks (7-6 or 6-7 score)
        if (w_set == 7 and l_set == 6):
            w_tiebreaks_w += 1
        if (w_set == 6 and l_set == 7):
            l_tiebreaks_w += 1

    # Calculate the total number of tiebreaks
    tiebreaks = w_tiebreaks_w + l_tiebreaks_w
    total_games = w_games + l_games

    return pd.Series([w_games, l_games, total_games, w_tiebreaks_w, l_tiebreaks_w, tiebreaks])

# Apply the function to each row in the 'score' column
data_atp[['w_games', 'l_games', 'total_games', 'p1_tiebreaks_w', 'p2_tiebreaks_w', 'tiebreaks']] = data_atp['score'].apply(calculate_games)
```
